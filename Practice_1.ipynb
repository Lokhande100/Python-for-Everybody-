```python
#10. Write a Python program that accepts an integer (n) and computes the value of n+nn+nnn. 
#Sample value of n is 5

#Expected Result : 615
```

```python
num = int(input("Enter an int value : "))
n1 = int( "%i" % num )
n2 = int( "%i%i" % (num,num) )
n3 = int( "%i%i%i" % (num,num,num) )
print (n1+n2+n3)
```
