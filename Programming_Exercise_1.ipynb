23. Write a Python program to get the n (non-negative integer) copies of the first 2 characters of a given string. 
Return the n copies of the whole string if the length is less than 2. 

```python
def substring_copy(str, n):
    flen = 2
    if flen > len(str):
        flen = len(str)
    substr = str[:flen]
  
    result = ""
    for i in range(n):
        result = result + substr
    return result

print(substring_copy('abcdef', 2))
print(substring_copy('nr', 2))
```

32. Write a Python program to get the least common multiple (LCM) of two positive integers.

```python
def lcm(x, y):
    if x > y:
        z = x
    else:
        z = y
    while(True):
        if((z % x == 0) and (z % y == 0)):
            lcm = z
            break
        z += 1
    return lcm
```

```python
print (lcm(5,8))
```

```python
def max_min(number):
    l = number[0]
    s = number[0]
    for num in number:
        if num> l:
            l = num
        elif num< s:
            s = num
    return l, s

print(max_min([1,2,3,-5,-8,45]))
```

149. Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number. 

```python
def sum_of_cubes(n):
    n=n-1
    total = 0
    while n > 0:
        total=total + n * n * n
        n=n-1
        return total
    
    
print("Sum of cubes: ",sum_of_cubes(9))
```
